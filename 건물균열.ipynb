```python
# 사용자가 제시한 조건에 맞는 파일명을 리스트 형식으로 변환하는 glob 사용
%cd /
from glob import glob

img_list = glob('/content/drive/MyDrive/building/images/*.tiff')

print(len(img_list))
```

```python
# 검증데이터와 학습데이터 분할
from sklearn.model_selection import train_test_split

train_img_list, val_img_list = train_test_split(img_list, test_size=0.2, random_state=2000)

print(len(train_img_list), len(val_img_list))
```

```python
# 파일 이름 수정
with open('/content/drive/MyDrive/building/train.txt', 'w') as f:
    f.write('\n'.join(train_img_list) + '\n')
with open('/content/drive/MyDrive/building/val.txt', 'w') as f:
    f.write('\n'.join(val_img_list) + '\n')
```

```python
import yaml
```

```python
# yaml파일을 파싱해서 python 객체로 불러오기
with open('/content/drive/MyDrive/building/images/data.yaml', 'r')  as f:
    data = yaml.full_load(f)

print(data)

data['train'] = '/content/drive/MyDrive/building/train.txt'
data['val'] = '/content/drive/MyDrive/building/val.txt'

with open('/content/drive/MyDrive/building/images/data.yaml', 'w') as f:
    yaml.dump(data, f)

print(data)
```
