```python
import os
import cv2
from glob import glob
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
img_list = glob('/content/drive/MyDrive/골프영상/주행영상_5/*.MP4')
```

```python
len(img_list)
```

```python
img_list
```

```python
from glob import glob
count = 0
for i in range(0,8):
    vidcap1 = cv2.VideoCapture(img_list[i])
    os.mkdir("/content/drive/MyDrive/골프영상/주행영상_5/" + str(i))
    while(vidcap1.isOpened()):
        ret, image = vidcap1.read()
        if not ret:
            break
        image = cv2.resize(image, (960,540))

        if(int(vidcap1.get(1)) % 3 == 0):
            print('Saved frame number :' + str(int(vidcap1.get(1))))
            cv2.imwrite("/content/drive/MyDrive/골프영상/주행영상_5/"+str(i)+"/frame%d.jpg" % count, image)
            count += 1

    vidcap1.release()
```
